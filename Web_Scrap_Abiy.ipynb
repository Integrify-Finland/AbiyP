# General introduction
This work is done to scrape data from a Swedish online shopping wibsite called blocket.se particularily about cars that are listed for sale. The main objective is to extract information regarding the price, brand name, mileage, the location, gear type and fuel type of each car listed.

# Building a dataframe from the extracted data features 

```python
li_1=[]

for item in car_blc:
    dic_1={}
    city=item.find('div',{'class':'crmFWf'})
    city=item.find_all('p')[0].find_all('a')[1].text
    dic_1['city']=city
    
    brand=item.find('div',{'class':'leTJeS'}).span.text
    dic_1['car_brand']=brand
    
    year=item.find_all('li')[0].text
    dic_1['Manufacture_year']=year
    
    Fuel_Type=item.find_all('li')[1].text
    dic_1['Fuel_type']=Fuel_Type
    
    Mileage=item.find_all('li')[2].text
    dic_1['Mileage']=Mileage
    
    Gear_type=item.find_all('li')[3].text
    dic_1['Gear_type']=Gear_type
    
    price=item.find('div',{'class':'jVOeSj'}).text
    dic_1['Price']=price
    
        
    li_1.append(dic_1)
  
li_1

    
    
```

```python
# remove units from columns
# check the car brand and try to see if there is extra info and try to remove unnecessary ones using regex
# wget
```

```python
df_1=pd.DataFrame(li_1)
df_1
```

# Improving the dataframe with regex

```python
import re
```

```python
df_1['Price']=df_1.Price.apply(lambda x: int(re.sub('\s','',x).strip('kr')))
df_1
```

```python
li_mil=[]
for i in df_1.Mileage:
    try:
        m=i.split('-')[1].strip('mil')
        m=re.sub('\s','',m)
        li_mil.append(m)
    except:
        m=i.split('-')[0].strip('mil')
        m=re.sub('\s','',m)
        li_mil.append(m)
        
li_mil
#df_1.Mileage[3].split('-')[1]
```
